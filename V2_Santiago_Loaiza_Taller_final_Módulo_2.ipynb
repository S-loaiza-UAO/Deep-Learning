### Que es el procesamiento de lenguaje natural (NLP)

###Fundamentacion para la arquitectura de un Transformer [[4]](https://arxiv.org/pdf/1706.03762)

Basados en el documento "Attention Is All You Need", el modelo Transformer, es una arquitectura novedosa para el procesamiento de lenguaje natural que se basa completamente en mecanismos de atención.

A diferencia de los modelos recurrentes o convolucionales, el Transformer evita la computación secuencial, lo que permite un mayor paralelismo y una mayor eficiencia en el entrenamiento.

**¿Que oportunidades motivaron este nuevo modelo?**

* Limitaciones de modelos recurrentes: Los modelos recurrentes procesan la información secuencialmente, lo que dificulta el paralelismo y la eficiencia en el manejo de secuencias largas.
* Importancia de la atención: Los mecanismos de atención han demostrado ser cruciales para modelar dependencias a larga distancia en el lenguaje.
* Ventajas del Transformer: El Transformer ofrece un paralelismo significativamente mayor, lo que reduce el tiempo de entrenamiento y mejora la calidad de la traducción.


**Arquitectura del Modelo:**

El Transformer sigue una estructura de codificador-decodificador. Tanto el codificador como el decodificador están compuestos por una pila de capas idénticas que incluyen:
* Atención multi-cabeza: Permite al modelo atender a información de diferentes subespacios de representación en diferentes posiciones de la secuencia.
* Redes de alimentación hacia adelante: Aplican transformaciones lineales con una función de activación ReLU a cada posición de la secuencia.
* Codificación posicional: Inyecta información sobre la posición de los tokens en la secuencia, ya que el modelo no tiene en cuenta el orden inherente de las palabras.

image.png

**Tipos de Atención:**

* Atención codificador-decodificador: Permite que cada posición en el decodificador atienda a todas las posiciones en la secuencia de entrada.
* Auto-atención del codificador: Permite que cada posición en el codificador atienda a todas las posiciones en la capa anterior del codificador.
* Auto-atención del decodificador: Permite que cada posición en el decodificador atienda a todas las posiciones anteriores en el decodificador, manteniendo la propiedad auto-regresiva.


image.png

**Resultados:**

* Traducción automática: El Transformer supera el rendimiento de modelos anteriores en las tareas de traducción inglés-alemán e inglés-francés, logrando un nuevo estado del arte en calidad de traducción.
* Análisis sintáctico: El Transformer también se aplica con éxito al análisis sintáctico del inglés, demostrando su capacidad de generalización a otras tareas de PNL.


**Ventajas del Transformer:**

* Paralelismo: Permite un entrenamiento más rápido y eficiente.
* Eficiencia: Requiere menos tiempo de entrenamiento en comparación con los modelos recurrentes y convolucionales.
* Calidad: Alcanza un nuevo estado del arte en calidad de traducción.
* Generalización: Se puede aplicar con éxito a otras tareas de PNL.


**Conclusion:**

El Transformer representa un avance significativo en el procesamiento de lenguaje natural. Su arquitectura basada en la atención permite un paralelismo y una eficiencia sin precedentes, abriendo nuevas posibilidades para el desarrollo de modelos de lenguaje más potentes y versátiles.

####Probando ChatGPT 3.5 como chatbot

```python
def display_chat_history(messages):
    for message in messages:
        print(f"{message['role'].capitalize()}: {message['content']}")
```

```python
def get_assistant_response(messages):
    r = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[{"role": m["role"], "content": m["content"]} for m in messages],

    )
    response = r.choices[0].message.content
    return response
```

```python
messages = [{"role": "assistant", "content": "How can I help?"}]
```

```python
while True:
    display_chat_history(messages)

    user_input = input("User: ")
    messages.append({"role": "user", "content": user_input})

    assistant_response = get_assistant_response(messages)
    messages.append({"role": "assistant", "content": assistant_response})
```

```python
import torch
from transformers import DistilBertTokenizer, DistilBertForSequenceClassification

tokenizer = DistilBertTokenizer.from_pretrained("distilbert-base-uncased-finetuned-sst-2-english")
model = DistilBertForSequenceClassification.from_pretrained("distilbert-base-uncased-finetuned-sst-2-english")
```

```python
# Clasificar cada comentario usando "text-classification" de Hugging Face
clasificaciones3 = []
for comentario in comentarios_aleatorios["comentario"]:
  print(comentario)

  inputs = tokenizer("{comentario}", return_tensors="pt")
  with torch.no_grad():
    logits = model(**inputs).logits

  predicted_class_id = logits.argmax().item()


  # Generar la solicitud para el modelo
  response=  model.config.id2label[predicted_class_id]
  print(response)

  # Almacenar la clasificación
  clasificaciones3.append(response)

# Agregar la clasificación a cada comentario en el DataFrame
comentarios_aleatorios["clasificacion"] = clasificaciones3
```

```python
# Clasificar cada comentario usando "Zero Shot" de Hugging Face
likert_scale = ["Muy buena", "Buena", "Neutro", "Malo", "Muy malo"]

classzero = []
for comentario in comentarios_aleatorios["comentario"]:
  print(comentario)

  #1. Clasificar en las categorías originales
  sequence_to_classify = comentario
  candidate_labels =  ["Restaurant", "Location", "Bar", "Events", "customer service"]
  category_response= classifierZS(sequence_to_classify, candidate_labels, multi_label=True) #  multi_label=True / pra evaluar cada categoria ya que varias categorias pueden ser correctas.

  #2. Evaluar cada categoría con la escala Likert
  likert_classifications = {}
  for label, score in zip(category_response['labels'], category_response['scores']):
      # Si la puntuación es mayor a un umbral (por ejemplo, 0.5), evaluar con la escala Likert
      if score > 0.5:
          sequence_to_classify = f"{comentario} {label}"  # Combinar comentario con la categoría
          likert_response = classifierZS(sequence_to_classify, likert_scale)
          likert_classifications[label] = likert_response['labels'][0]
      else:
          likert_classifications[label] = "N/A"  # No aplica si la puntuación es baja

  print(f"Clasificaciones Likert: {likert_classifications}")

  #3.  Almacenar la clasificación
  classzero.append(likert_classifications)

# Agregar la clasificación a cada comentario en el DataFrame
comentarios_aleatorios["clasification"] = classzero
```
